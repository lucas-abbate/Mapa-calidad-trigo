## Mapa calidad trigo - 1 anio

```python
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.rcParams['figure.figsize'] = [12, 12]

gdf = pd.read_csv('Mapa calidad trigo 1 anio.csv')
gdf['Lat.'] = pd.to_numeric(gdf['Latitud'])
gdf['Long.'] = pd.to_numeric(gdf['Longitud'], downcast='integer')


lats = list(gdf['Latitud'])
lons = list(gdf['Longitud'])
group = list(gdf['Grupo'])
dpto = list(gdf['Dpto Label'].str.strip())
var1 = list(gdf['BON_media'])
var2 = list(gdf['ESTAB_media'])
var3 = list(gdf['GH_media'])
var4 = list(gdf['PH_media'])
var5 = list(gdf['PROT_media'])
var6 = list(gdf['RTO_media'])
var7 = list(gdf['P.BLANCA_media'])
var8 = list(gdf['W_media'])


d = dict(
    lat = lats,     #0
    lon = lons,     #1
    group = group,  #2
    dpto = dpto,    #3
    var1 = var1,    #4
    var2 = var2,    #5
    var3 = var3,    #6
    var4 = var4,    #7
    var5 = var5,    #8
    var6 = var6,    #9
    var7 = var7,    #10
    var8 = var8,    #11

)
df = pd.DataFrame(d)
gb = df.groupby(['group'])
group_name = list(set(df['group']))

names = ['',
'< 90',
'>90 a 92',
'>92 a 94',
'>94 a 96',
'>96 a 98',
'>98 a 100',
'>100 a 102',
'>102 a 104',
'>104 a 106',
'>106 a 108',
'>108 a 110',
'>110',
]

colscale = [ '', # Colores tomados de https://colordesigner.io/gradient-generator
'#fafa6e',
'#ffdf60',
'#ffc559',
'#fdab58',
'#f5935a',
'#e87c5d',
'#d76861',
'#c25763',
'#aa4864',
'#903c63',
'#74325e',
'#582a57'
]

mapbox_access_token = 'pk.eyJ1IjoiYmx1emVyMzMiLCJhIjoiY2tzZG1lazh0MHJwbDJvcDRmcjczcm45NSJ9.ukPsicfdGZGdCBlbo9U8dw'

fig = go.Figure()

for gr in group_name:
    my_gr = gb.get_group(gr)

    fig.add_trace(go.Scattermapbox(
    lon=my_gr['lon'],
    lat=my_gr['lat'],
    mode='markers',
    marker=go.scattermapbox.Marker(
        size=13,
        color= colscale[gr]
    ),
    text=my_gr['dpto'],
    name=names[gr],
    customdata = my_gr,
    hovertemplate =
        "%{customdata[3]}<br>" +
        "   Bonificación comercial (%):                      %{customdata[4]}<br>" +
        "   Estabilidad farinográfica (min):                 %{customdata[5]}<br>" +
        "   Gluten húmedo de la harina (%, humedad 13.5 %): %{customdata[6]}<br>" +
        "   Granos panza blanca (%):                        %{customdata[10]}<br>" +
        "   Peso hectolítrico (kg/hl):                      %{customdata[7]}<br>" +
        "   Proteína del grano (%, humedad 13.5 %):         %{customdata[8]}<br>" +
        "   Rendimiento en grano (kg/ha):                   %{customdata[9]}<br>" +
        "   W alveográfico (J/10^4):                         %{customdata[11]}<br>" +
        "<extra></extra>"
))


fig.update_layout(
    legend_bgcolor = '#ADD8E6',
    legend_title = 'Bonificación comercial (%)',
    hoverlabel=dict(
        font_size=12,
        font_family="Space Mono, Droid Sans Mono, Courier new"
    ),
    hovermode='closest',
        margin=dict(
        l=0,
        r=0,
        b=0,
        t=0, #PEA: Con título usar 40
        pad=4),
    paper_bgcolor='#ADD8E6',
    legend=dict(
        y=.96
    ),
    mapbox=dict(
        accesstoken=mapbox_access_token,
        bearing=0,
        center=dict(
            lat=-35,
            lon=-62,
        ),
        pitch=0,
        zoom=4.5, #4,
        style='mapbox://styles/bluzer33/cksdm1t080syw17nrekv0ltho'
    ),
)
fig.show()
fig.write_html('Mapa calidad trigo 1 anio.html')
```
